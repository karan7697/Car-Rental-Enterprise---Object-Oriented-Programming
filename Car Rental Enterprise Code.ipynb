System For Car Rental Enterprise:
As we know alot of people cannot afford their private vehicles and in this case they turn themselves to Rental Car Enterprises. I have designed a system which will help the rentals to keep track of their inventory and also ensure a log of cars rented out. 

```python
import datetime

class Car_Rental:    #Defining the main class
    def __init__(self, listofCars):  
        self.cars = listofCars

    def displayAvailableCars(self):   #Function to Display a list of available cars in the inventory
        print(f"\n{len(self.cars)} Cars In Inventory: ")
        for car in self.cars:
            print(" ♦-- " + car)
        print("\n")

    def borrowCar(self, name, carname): #Function to rent or issue a car to a customer
        if carname not in self.cars:
            print(
                f"{carname} Car is issued to someone else. Please wait until returned.\n") #If car is already issues to another user
        else:
            track.append({name: carname})
            print("CAR ISSUED : Thankyou. Have a wonderful Drive and return on time.\n")
            self.cars.remove(carname)

    def returnCar(self, carname): #Function to return the car and append it to the inventory again
        print("CAR RETURNED : THANK YOU. HOPE YOU HAD A WONDERFUL! \n")
        self.cars.append(carname)

    def SortingByAge(Data): #Function to sort the age of the cars
      size = len(Data)
      for i in range(size):
          min_index = i
          for j in range(i + 1, size):
              if Data[min_index]["Age"] > Data[j]["Age"]:
                  min_index = j
          Data[i], Data[min_index] = Data[min_index], Data[i]

      print(Data)
```

# Time Complexity:
The time complexity of the sorting algorithm used in the "SortingByAge" is **Quadratic time complexity - O(n^2)**. 
This algorithm is said to have a quadratic time complexity because it needs to perform a linear time operation for each value in the input data. For Example: it need to compare the age of each car to all other ages of the cars present in the list.

# Space Complexity:
Space complexity remains O(1) as we are not using any excess memory

# Computational Complexity of Searching Algorthms:
Computational  complexity for the searching algorithm is O(1)+O(n^2) . Therefore, the operation that requires O(n^2) will be the bottleneck of this algorithm as the size of the input data increases. Based on this, we may say that this algorithm's temporal complexity is O(n^2).
